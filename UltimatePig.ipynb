```python
from pigsolitaire import *
import statistics as st
import random
import matplotlib.pyplot as plt

%matplotlib inline

# Player only chooses once
count = 1
result_list = []
while count <= 10000:
    player1 = Player("ENIAC")
    game = Game(player1)
    game.start()
    result_list.append(player1.score)
    count += 1

mean = st.mean(result_list)
stdev = st.pstdev(result_list)
print("Mean is: {}".format(mean))
print("Standard Deviation is: {}".format(stdev))

# Line Plot
plt.xlabel("Trial Number")
plt.ylabel("Total Points")
plt.title("Total Points Every Trial")
plt.plot(result_list)
plt.show()

# Histogram
plt.hist(result_list, bins=50)
plt.xlabel("Trial Number")
plt.ylabel("Total Points")
plt.title("Total Points Every Trial")
ymin, ymax = plt.ylim()
plt.vlines(mean, ymin, ymax, color="blue")
plt.vlines([mean - 2* stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, 
           linestyles="dashed", color="red")
```
